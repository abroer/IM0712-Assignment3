```python
from xml.sax.handler import feature_string_interning

# used libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Dense, Input
from sklearn.metrics import mean_squared_error, mean_absolute_percentage_error
from scipy.stats import pearsonr
from skopt import BayesSearchCV
from skopt.space import Real, Integer
from scikeras.wrappers import KerasRegressor
from sklearn.pipeline import Pipeline
from tensorflow.keras.optimizers import Adam
from sklearn.model_selection import GridSearchCV
```

```python
# #import the winequality dataset

# # Define the relative path to the CSV file
# relative_path = 'wine+quality/winequality-white.csv'
# Read the CSV file using pandas
wine_df = pd.read_csv("/lakehouse/default/Files/winequality-white.csv", delimiter=';')

features = wine_df.drop('quality', axis=1)
target = wine_df['quality']

len(features)
```

Feature scaling was applied using standardization. Following code standardizes the dataset.

```python
scaler = StandardScaler()

features_scaled = scaler.fit_transform(features)

feature_scaled_df = pd.DataFrame(features_scaled, columns=features.columns)
```

Create the train, validation and test set from the original data set

```python
# Step 1: Split the data into training and temporary sets
# The training set will contain 60% of the data, and the temporary set will contain 40% of the data
features_train, features_temp, target_train, target_temp = train_test_split(features_scaled, target, test_size=0.4, random_state=1)

# Step 1: Split the data into training and temporary sets
# The training set will contain 60% of the data, and the temporary set will contain 40% of the data
features_test, features_validate, target_test, target_validate = train_test_split(features_temp, target_temp, test_size=0.5, random_state=1)

# Now, features_train and target_train contain the training data (60% of the original data)
# features_test and target_test contain the testing data (20% of the original data)
# features_validate and target_validate contain the validation data (20% of the original data)
```

High level search for a better number of neurons

```python

# Define a function to create the model, required for KerasRegressor
def create_model(neurons=15, hidden_layers=1):
    model = Sequential()  # Initialize the Sequential model
    model.add(Input(shape=(11,)))  # Add the input layer with 11 features
    
    # Add the specified number of hidden layers
    for _ in range(hidden_layers):
        model.add(Dense(neurons, activation='relu'))  # Add hidden layers with 'neurons' neurons and ReLU activation

    model.add(Dense(1, activation='linear'))  # Add the output layer with 1 neuron and linear activation
    model.compile(loss='mse', optimizer='adam', metrics=['mean_squared_error'])  # Compile the model with MSE loss and Adam optimizer
    return model  # Return the compiled model

# Wrap the model using KerasRegressor
model = KerasRegressor(model=create_model, verbose=3)

# Define the parameter grid to search over
param_grid = {
    'batch_size': [18],  # Different batch sizes to try
    'epochs': [50],  # Different numbers of epochs to try, from 50 to 150 in steps of 10
    'model__neurons': list(range(10, 61, 5)),  # Different numbers of neurons to try in each hidden layer
    'model__hidden_layers': [3]  # Different numbers of hidden layers to try
}

# Set up GridSearchCV with the model, parameter grid, 3-fold cross-validation, and MSE scoring
grid = GridSearchCV(estimator=model, param_grid=param_grid, cv=3, scoring='neg_mean_squared_error', verbose=3)

# Fit the GridSearchCV to the training data
grid_result = grid.fit(features_train, target_train)

# Print the best score and the best parameters found by GridSearchCV
print(f"Best: {grid_result.best_score_} using {grid_result.best_params_}")

# Retrieve the best model found by GridSearchCV
best_model = grid_result.best_estimator_

# Train the best model on the training data and validate on the validation data
history = best_model.fit(features_train, target_train, validation_data=(features_validate, target_validate))
```

```python

# Extract the results from the BayesSearchCV object
results = pd.DataFrame(grid_result.cv_results_)
```

```python
# Plot the impact of 'batch_size' on the mean test score using scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(results['param_model__neurons'], results['mean_test_score'], marker='o')
plt.xlabel('model__neurons')
plt.ylabel('Mean Test Score')
plt.title('Impact of model__neurons on Mean Test Score')
plt.grid(True)
plt.show()
```
